## Cuánto es λ?
Si el promedio de arribos es de **1 avión por hora**, podemos calcular λ.

- Cada hora tiene 60 minutos.
- La variable de arribos por minuto sigue una distribución Bernoulli con probabilidad λ.
- La suma en 60 minutos es una Binomial con parámetros n = 60 y p = λ.
- La esperanza es: E[X] = n · p = 60 · λ.

Queremos que E[X] = 1.  
Entonces:

**60 · λ = 1 → λ = 1/60 ≈ 0.0167**

Esto significa que en cada minuto hay una probabilidad de **1.67%** de que aparezca
un avión en el radar.  
En un día de 18 horas (1080 minutos), la esperanza teórica de arribos es:

**1080 · λ = 18 arribos**


```python
from fractions import Fraction

promedio_deseado = 1   # 1 avión por hora
n_min = 60             # minutos en una hora

lam = Fraction(promedio_deseado, n_min)

print(f"Si el promedio es {promedio_deseado} avión/hora:")
print(f"λ = {lam} (probabilidad de arribo por minuto)")
```
